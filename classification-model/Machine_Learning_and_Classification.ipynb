```python
#features
#long
#short leg
#makes "au"

pig_1 = [0,1,0]
pig_2 = [0,1,1]
pig_3 = [1,1,0]

dog_1 = [0,1,1]
dog_2 = [1,0,1]
dog_3 = [1,1,1] 

# pig => 1,  dog => 0
train_x = [pig_1, pig_2, pig_3, dog_1, dog_2, dog_3]
train_y = [1,1,1,0,0,0]
```

```python
from sklearn.svm import LinearSVC

model = LinearSVC()
model.fit(train_x, train_y)
```

```python
animal_1 = [1,1,1]
model.predict([animal_1])
```

```python
animal_1 = [1,1,1]
animal_2 = [1,1,0]
animal_3 = [0,1,1]

tests_x = [animal_1, animal_2, animal_3]
tests_y = [0,1,1]

previews = model.predict(tests_x)
```

```python
corrects = (previews == tests_y).sum()
total = len(tests_x)
accuracy = corrects/total

print("correct tax: %.2f" % (accuracy * 100))
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(tests_y, previews)*100
print("correct tax: %.2f" % accuracy)
```
